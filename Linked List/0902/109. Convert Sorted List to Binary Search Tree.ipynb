    Input: head = [-10,-3,0,5,9]
    Output: [0,-3,9,-10,null,5]
    Explanation: One possible answer is [0,-3,9,-10,null,5], which represents the shown height balanced BST.

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution:
    def sortedListToBST(self, head: ListNode) -> TreeNode:
        if not head:
            return head
        if not head.next:
            return TreeNode(head.val)
        
        slow = head
        fast = head.next
        prev = None
        while fast and fast.next:
            prev = slow
            slow = slow.next
            fast = fast.next.next
        
        if prev:
            prev.next = None
        else:
            head = None
        node = TreeNode(slow.val)
        node.left = self.sortedListToBST(head)
        node.right = self.sortedListToBST(slow.next)
        return node
```

```python
class Solution:
    def sortedListToBST(self, head: ListNode) -> TreeNode:
        if not head:
            return head
        if not head.next:
            return TreeNode(head.val)
        self.vals = []
        self.helper(head)
        m_idx = len(self.vals) // 2
        t_node = TreeNode(self.vals[m_idx])
        self.build(t_node, self.vals[:m_idx], self.vals[m_idx+1:])
        return t_node
    
    def build(self, root, left, right):
        if left:
            m_idx = len(left) // 2
            root.left = TreeNode(left[m_idx])
            self.build(root.left, left[:m_idx], left[m_idx+1:])
        if right:
            m_idx = len(right) // 2
            root.right = TreeNode(right[m_idx])
            self.build(root.right, right[:m_idx], right[m_idx+1:])
        
    
    def helper(self, root):
        if root:
            self.vals.append(root.val)
            self.helper(root.next)
```

```python
# 快慢指针
class Solution:
    def sortedListToBST(self, head: ListNode) -> TreeNode:
        def list_find(start, end):
            slow = fast = start
            while fast != end and fast.next != end:
                slow = slow.next
                fast = fast.next.next
            return slow
        
        def build_tree(start, end):
            if start == end:
                return
            mid = self.list_find(start, end)
            tree = TreeNode(mid.val)
            tree.left = build_tree(start, mid)
            tree.right = build_tree(mid.next, end)
            return tree
        return build_tree(head, None)
        
```
