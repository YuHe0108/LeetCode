```python
from collections import deque
import copy

class Solution:
    def slidingPuzzle(self, board):
        def encode(mat):
            res = ''
            for i in range(rows):
                for j in range(cols):
                    res += str(mat[i][j])
            return res
        
        rows, cols = 2, 3
        target = encode([[1, 2, 3], [4, 5, 0]])
        dq = deque([[board, 0]])
        seen = {encode(board)}
        while dq:
            for _ in range(len(dq)):
                temp, step = dq.popleft()
                temp = copy.deepcopy(temp)
                if encode(temp) == target:
                    return step
                
                for r in range(rows):
                    for c in range(cols):
                        if temp[r][c] == 0:
                            for nr, nc in (r+1, c), (r-1, c), (r, c+1), (r, c-1):
                                if 0 <= nr < rows and 0 <= nc < cols:
                                    temp[nr][nc], temp[r][c] = temp[r][c], temp[nr][nc]
                                    code_temp = encode(temp)
                                    if code_temp not in seen:
                                        seen.add(code_temp)
                                        dq.append((copy.deepcopy(temp), step + 1))
                                    temp[nr][nc], temp[r][c] = temp[r][c], temp[nr][nc]
        return -1                   
```

```python
solution = Solution()
solution.slidingPuzzle([[3,2,4],[1,5,0]])
```

```python
class Solution:
    def slidingPuzzle(self, board) -> int:
        moves = {0: {1, 3}, 1:{0, 2, 4}, 2:{1, 5}, 
                 3:{0, 4}, 4:{1, 3, 5}, 5:{2, 4}}
        used, cnt = set(), 0
        s = "".join(str(c) for row in board for c in row)
        q = [(s, s.index("0"))]
        while q:
            new = []
            for s, i in q:
                used.add(s)
                if s == "123450":
                    return cnt
                arr = [c for c in s]
                for move in moves[i]:
                    new_arr = arr[:]
                    new_arr[i], new_arr[move] = new_arr[move], new_arr[i]
                    new_s = "".join(new_arr)
                    if new_s not in used:
                        new.append((new_s, move))
            cnt += 1
            q = new
        return -1
```

```python
solution = Solution()
solution.slidingPuzzle([[3,2,4],[1,5,0]])
```
