```python
from collections import defaultdict

class Solution:
    def maxNumEdgesToRemove(self, n: int, edges) -> int:
        def find(x, father):
            if father[x] != x:
                father[x] = find(father[x], father)
            return father[x]
        
        def union(x, y, father):
            x, y = find(x, father), find(y, father)
            if x < y:
                father[y] = x
            else:
                father[x] = y
        
        def valid(count):
            for v in count.values():
                if len(v) == n:
                    return True
            return False
    
        N, useful = len(edges), 0
        all_father = [x for x in range(n + 1)]
        all_can = [x for x in edges if x[0] == 3]
        all_count = defaultdict(set)
        for _, a, b in all_can:
            if find(a, all_father) != find(b, all_father):
                useful += 1
                union(a, b, all_father)
                f = find(a, all_father)
                all_count[f].add(a)
                all_count[f].add(b)
                for v in all_count.values():
                    if len(v) == n:
                        return N - useful
    
        a_can = [x for x in edges if x[0] == 1]
        a_father = [x for x in all_father]
        a_count = {k:v for k, v in all_count.items()}
        a_can_break = False
        for _, a, b in a_can:
            if find(a, a_father) != find(b, a_father):
                useful += 1
                union(a, b, a_father)
                f = find(a, a_father)
                if a in a_count:
                    a_count[f] |= a_count[a]
                if b in a_count:
                    a_count[f] |= a_count[b]
                for v in a_count.values():
                    if len(v) == n:
                        a_can_break = True
                        break
            if a_can_break:
                break
        if not valid(a_count):
            return -1
        
        b_can = [x for x in edges if x[0] == 2]
        b_father = [x for x in all_father]
        b_count = {k:v for k, v in all_count.items()}
        b_can_break = False
        for _, a, b in b_can:
            if find(a, b_father) != find(b, b_father):
                useful += 1
                union(a, b, b_father)
                f = find(a, b_father)
                if a in b_count:
                    b_count[f] |= b_count[a]
                if b in a_count:
                    b_count[f] |= b_count[b]
                if valid(b_count):
                    b_can_break = True
                    break
            if b_can_break:
                break
        if valid(b_count) == False:
            return -1
        
        return N - useful
```

```python
from collections import defaultdict

class Solution:
    def maxNumEdgesToRemove(self, n: int, edges) -> int:
        def find(x, father):
            if father[x] != x:
                father[x] = find(father[x], father)
            return father[x]
        
        def union(x, y, father):
            x, y = find(x, father), find(y, father)
            if x < y:
                father[y] = x
            else:
                father[x] = y

        a_paths, b_paths, ab_paths = [], [], []
        for c, a, b in edges:
            if c == 1:
                a_paths.append((a, b))
            elif c == 2:
                b_paths.append((a, b))   
            else:
                ab_paths.append((a, b))
        
        ab_father = [x for x in range(n + 1)]
        a_count = b_count = ab_count = 0
        for a, b in ab_paths:
            f_a, f_b = find(a, ab_father), find(b, ab_father)
            if f_a != f_b:
                ab_count += 1
                union(a, b, ab_father)
        if ab_count == n:
            return len(edges) - ab_count
        
        a_father = [x for x in ab_father]
        for a, b in a_paths:
            f_a, f_b = find(a, a_father), find(b, a_father)
            if f_a != f_b:
                a_count += 1
                union(a, b, a_father)
        for i in range(1, n+1):
            if find(1, a_father) != find(i, a_father):
                return -1
        
        b_father = [x for x in ab_father]
        for a, b in b_paths:
            f_a, f_b = find(a, b_father), find(b, b_father)
            if f_a != f_b:
                b_count += 1
                union(a, b, b_father)
        for i in range(1, n+1):
            if find(1, b_father) != find(i, b_father):
                return -1
        return len(edges) - a_count - b_count - ab_count
```
