```python
class Node:
    def __init__(self, key, val, nxt=None):
        self.key = key
        self.val = val
        self.next = nxt


class MyHashMap:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.key_space = 1001
        self.tbl = [Node(None, None, None) for _ in range(self.key_space)]

    def put(self, key: int, value: int) -> None:
        """
        value will always be non-negative.
        """
        hash_key = key % self.key_space
        prv = self.tbl[hash_key]
        p = prv.next
        while p:
            if p.key == key:
                p.val = value
                return
            prv, p = p, p.next
        prv.next = Node(key, value)

    def get(self, key: int) -> int:
        """
        Returns the value to which the specified key is mapped, or -1 if this map contains no mapping for the key
        """
        hash_key = key % self.key_space
        p = self.tbl[hash_key].next
        while p:
            if p.key == key:
                return p.val
            p = p.next
        return -1

    def remove(self, key: int) -> None:
        """
        Removes the mapping of the specified value key if this map contains a mapping for the key
        """
        hash_key = key % self.key_space
        prv = self.tbl[hash_key]
        p = prv.next
        while p:
            if p.key == key:
                prv.next = p.next
                return
            prv, p = p, p.next
```

```python
# 使用链表来存储数据
class Node:
    def __init__(self, key, value, nxt=None):
        self.key = key # 存储了键值对
        self.value = value
        self.nxt = nxt

class MyHashMap:
    def __init__(self):
        self.size = 1000 # 定义map的总长度
        self.hash_map = [None] * self.size
    
    def _hash(self, key): # 获取 key hash化之后的值，这个值是唯一的
        return hash(key) % self.size 

    def put(self, key: int, value: int) -> None:
        hash_key = self._hash(key) # 获得 key 的哈希值
        if self.hash_map[hash_key] is None:
            self.hash_map[hash_key] = Node(key, value)
        else:
            prev_node = None
            cur_node = self.hash_map[hash_key] 
            while cur_node:
                if cur_node.key == key: # 更新值
                    cur_node.value = value
                    return
                prev_node = cur_node
                cur_node = cur_node.nxt
            prev_node.nxt = Node(key, value) # 将当前的值拼接在链表的最后
            
    def get(self, key: int) -> int:
        hash_key = self._hash(key) # 获得 key 的哈希值
        if self.hash_map[hash_key] is None:
            return -1
        cur_node = self.hash_map[hash_key]
        while cur_node:
            if cur_node.key == key:
                return cur_node.value
            cur_node = cur_node.nxt
        return -1
        
    def remove(self, key: int) -> None:
        hash_key = self._hash(key) # 获得 key 的哈希值
        if self.hash_map[hash_key]:
            prev_node = None
            cur_node = self.hash_map[hash_key]
            while cur_node:
                if cur_node.key == key:
                    if prev_node:
                        prev_node.nxt = cur_node.nxt
                    else:
                        self.hash_map[hash_key] = cur_node.nxt
                prev_node = cur_node
                cur_node = cur_node.nxt
```

```python
hash_map = MyHashMap()
hash_map.put('a', 10)
```
