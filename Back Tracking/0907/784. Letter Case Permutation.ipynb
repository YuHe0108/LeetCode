```python
class Solution:
    def letterCasePermutation(self, S: str):
        res = [S]
        
        def dfs(s_arr, temp, count):
            if count == len(res[0]):
                res.append(''.join(temp[:]))
                return
            for i, s in enumerate(s_arr):
                if 65 <= ord(s) <= 90:
                    s = chr(ord(s) + 32) # 大写转换为小写
                elif 97 <= ord(s) <= 122:
                    s = chr(ord(s) - 32) # 小写转换为大写
                temp.append(s)
                dfs(s_arr[i+1:], temp, count+1)
                temp.pop()
        
        dfs(S, [], 0)
        return res
```

```python
class Solution:
    def letterCasePermutation(self, S: str):
        self.res = []
        self.dfs(S, '', 0)
        return self.res
    
    def dfs(self, S, temp, count):
        if count == len(S):
            self.res.append(temp)
            return
        if S[count].isalpha():
            self.dfs(S, temp+S[count].lower(), count+1)
            self.dfs(S, temp+S[count].upper(), count+1)
        else:
            self.dfs(S, temp+S[count], count+1)
```

```python
S_ = "sDYnJlPpsJ"
solution = Solution()
solution.letterCasePermutation(S_)
```
