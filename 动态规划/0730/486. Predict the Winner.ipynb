    给出一个非负整数的分数数组。
    玩家1从数组的任意一端选择一个数字，
    然后是玩家2，然后是玩家1，依此类推。
    每次玩家选择一个号码时，下一个玩家将无法使用该号码。
    直到所有的分数都被选中为止。得分最高的选手获胜。
    给出一组分数，预测玩家1是否是赢家。
    你可以假设每个玩家都是为了最大化他的得分。
    
    例1：
        Input: [1, 5, 2]
        Output: False
            Explanation: Initially, player 1 can choose between 1 and 2. 
            If he chooses 2 (or 1), then player 2 can choose from 1 (or 2) and 5. 
            If player 2 chooses 5, then player 1 will be left with 1 (or 2). 
            So, final score of player 1 is 1 + 2 = 3, and player 2 is 5. 
            Hence, player 1 will never be the winner and you need to return False.
    例2：
        Input: [1, 5, 233, 7]
        Output: True
        Explanation: Player 1 first chooses 1. 
        Then player 2 have to choose between 5 and 7. No matter which number player 2 choose, player 1 can choose 233.
        Finally, player 1 has more score (234) than player 2 (12), so you need to return True representing player1 can win.
        
    规则：
        1、每位玩家都只能从两头挑选数字
        2、两位玩家得分相同时，玩家1获胜

```python
class Solution:
    def PredictTheWinner(self, nums) -> bool:
        len_nums = len(nums)
        if len_nums <= 1: return True
        dp_1 = [0] * (len_nums // 2 if len_nums % 2 == 0 else len_nums // 2 + 1)
        dp_2 = [0] * (len_nums // 2)
        index = 0
        while len(nums) > 0:
            val_1, pop_1 = self.helper(nums)
            nums.pop(pop_1)
            dp_1[index] = val_1
            if len(nums) != 0:
                val_2, pop_2 = self.helper(nums)
                nums.pop(pop_2)
                dp_2[index] = val_2
            index += 1
        print(dp_1, dp_2)
        if sum(dp_1) >= sum(dp_2): return True
        return False
    
    def helper(self, inputs):
        num_len = len(inputs)
        if num_len == 1: return inputs[0], 0
        if num_len == 2: return max(inputs), inputs.index(max(inputs))
        if inputs[1] > inputs[-2]:
            return inputs[-1], -1
        elif inputs[1] == inputs[-2]:
            if inputs[-1] >= inputs[0]:
                return inputs[-1], -1
            else:
                return inputs[0], 0
        else:
            return inputs[0], 0
```

```python
class Solution:
    def PredictTheWinner(self, nums) -> bool:
        dp = [[0] * len(nums) for _ in range(len(nums))]
        
        # 当nums的长度位1时：表示的是 从第 i 个 到第 i 个， 玩家1的得分，很显然只能是num[i]
        for i in range(len(nums)):
            dp[i][i] = nums[i]
        # 当nums的长度位2时：如果只存在两个数，取其中最大值
        for i in range(len(nums)-1):
            dp[i][i+1] = max(nums[i], nums[i+1])
        
        
        # 当nums的长度在三个以上时
        for l in range(3, len(nums) + 1):
            for i in range(len(nums) + 1 - l):
                j = i + l - 1 # 每三个划分一段
                # 为什么时min，因为play——2一定会选择最大值，所以剩下的是最小值
                # dp[i+1][j-1]：玩家2拿了尾元素
                # dp[i+2][j]: 玩家2拿了首元素
                # 当玩家2开始选择的时候，因为要让玩家 2 拿到的分数尽可能的小
                head = min(dp[i+1][j-1], dp[i+2][j]) + nums[i]
                tail = min(dp[i][j-2], dp[i+1][j-1]) + nums[j]
                dp[i][j] = max(head, tail)
                
        print(dp)
        sum_play_1 = dp[0][-1]
        sum_nums = sum(nums)
        if sum_play_1 >= sum_nums - sum_play_1: return True
        return False
```
