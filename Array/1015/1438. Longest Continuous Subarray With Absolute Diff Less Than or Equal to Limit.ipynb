```python
class Solution:
    def longestSubarray(self, nums, limit: int) -> int:
        N = len(nums)
        if N == 1: return 1
        res = 1
        l_idx, r_idx = 0, 1
        min_val = min(nums[l_idx], nums[r_idx])
        max_val = max(nums[l_idx], nums[r_idx])
        while r_idx < N:
            if max_val - min_val <= limit:
                while r_idx < N and max_val - min_val < limit:
                    res = max(res, r_idx - l_idx + 1)
                    r_idx += 1
                    if r_idx < N:
                        min_val = min(min_val, nums[r_idx])
                        max_val = max(max_val, nums[r_idx])
            else:
                l_idx += 1
                r_idx += 1
                if l_idx < N and r_idx < N:
                    min_val = min(nums[l_idx], nums[r_idx])
                    max_val = max(nums[l_idx], nums[r_idx])
        return res
```

```python
import heapq

class Solution:
    def longestSubarray(self, nums, limit: int) -> int:
        N = len(nums)
        if N == 1: return 1
        res = 0
        r_idx = 0
        sub_arr =  [nums[0]]
        for l_idx in range(N):
            while heapq.nlargest(1, sub_arr)[0] - heapq.nsmallest(1, sub_arr)[0] <= limit:
                res = max(res, r_idx - l_idx + 1)
                r_idx += 1
                if r_idx < N:
                    heapq.heappush(sub_arr, nums[r_idx])
                else:
                    break
            if r_idx == N:
                break
            sub_arr.remove(nums[l_idx])
        return res
```

```python
from collections import deque

class Solution:
    def longestSubarray(self, nums, limit: int) -> int:
        N = len(nums)
        if N == 1: return 1
        res = 1
        r_idx = 0
        max_val = deque([0])
        min_val = deque([0])
        mx, mi = nums[0], nums[0]
        for l_idx in range(N):
            while mx - mi <= limit:
                res = max(res, r_idx - l_idx + 1)
                r_idx += 1
                if r_idx >= N:
                    break
                #维持一个单调deque
                while max_val and nums[r_idx] >= nums[max_val[-1]]:
                    max_val.pop()
                max_val.append(r_idx)
                mx = nums[max_val[0]]
                while min_val and nums[r_idx] <= nums[min_val[-1]]:
                    min_val.pop()
                min_val.append(r_idx)
                mi = nums[min_val[0]]
                
            if  r_idx >= N:
                break
            if max_val and max_val[0] <= l_idx:
                max_val.popleft()
                mx = nums[max_val[0]]
            if min_val and min_val[0] <= l_idx:
                min_val.popleft()
                mi = nums[min_val[0]]
        return res
```

```python
from collections import deque

class Solution:
    def longestSubarray(self, A, limit):
        N = len(A)
        if N == 1: return 1
        
        max_idxs = deque([0])
        min_idxs = deque([0])
        min_val, max_val = A[0], A[0]
        
        r_idx, ret = 0, 1
        for l_idx in range(N):
            while max_val - min_val <= limit:
                ret = max(ret, r_idx - l_idx + 1)
                r_idx += 1
                if r_idx >= N:
                    break
                
                # 维持两个单调栈
                while max_idxs and A[r_idx] >= A[max_idxs[-1]]:
                    max_idxs.pop()
                max_idxs.append(r_idx)
                max_val = A[max_idxs[0]]
                while min_idxs and A[l_idx] <= A[min_idxs[-1]]:
                    min_idxs.pop()
                min_idxs.append(r_idx)
                min_val = A[min_idxs[0]]
                
            if r_idx >= N:
                break
            if max_idxs and max_idxs[0] <= l_idx:
                max_idxs.popleft()
                max_val = A[max_idxs[0]]
            if min_idxs and min_idxs[0] <= l_idx:
                min_idxs.popleft()
                min_val = A[min_idxs[0]]
        return ret
```
