```python
from typing import List
from collections import defaultdict, deque

# 存在两层拓扑关系：
# 第一个是各个group之间存在关系，每队必须要等另一个队完成，自己才能开工
# 第二个是各个 item之间的拓扑关系，一个任务的完成依赖于其他任务

class Solution:
    def sortItems(self, n: int, m: int, group: List[int], beforeItems: List[List[int]]) -> List[int]:
        def topo_sort(items, indeg, nexts):
            outs = []
            points = deque()
            for i in items: # 遍历，只有入度为0，才能加入
                if indeg[i] == 0:
                    points.append(i)
                    outs.append(i)
                    
            while points:
                for _ in range(len(points)):
                    p = points.popleft()
                    for j in nexts[p]:
                        indeg[j] -= 1
                        if indeg[j] == 0:
                            points.append(j)
                            outs.append(j)
            return outs if len(outs) == len(items) else []
        
        # 每个小组有哪些任务
        group_ids = set()
        group_items = defaultdict(set)
        nexts_id = m
        for i, v in enumerate(group):
            if v == -1:
                group[i] = nexts_id
                nexts_id += 1
            group_items[group[i]].add(i)
            group_ids.add(group[i])
        
        # 首先进行组间的排序
        item_nexts = defaultdict(set) # 组内任务的依赖关系
        item_indeg = defaultdict(int) # 组内任务的入度
        for i in range(n):
            for item in beforeItems[i]:
                if group[item] == group[i]: # 只考虑小组之间的任务依赖关系
                    item_nexts[item].add(i)
                    item_indeg[i] += 1
        
        s_g_items = {} # 每个小队内部完成工作的先后顺序
        for g_id, g_item in group_items.items():
            out = topo_sort(g_item, item_indeg, item_nexts)
            if not out:
                return []
            s_g_items[g_id] = out 
        
        # 进行 group之间的排序
        group_nexts = defaultdict(set) # 小队依赖关系
        group_indeg = defaultdict(int) # 小队之间的入度关系
        for i in range(n):
            for j in beforeItems[i]:
                if group[j] != group[i]: # 只考虑不同小组之间的任务依赖关系
                    if group[i] not in group_nexts[group[j]]:
                        group_nexts[group[j]].add(group[i])
                        group_indeg[group[i]] += 1
        
        s_groups = topo_sort(group_ids, group_indeg, group_nexts) # 各个小队之间的完成先后顺序
        res = []
        for g_id in s_groups:
            res.extend(s_g_items[g_id])
        return res
```
