```python
from typing import List
import heapq

class Solution:
    def assignTasks(self, servers: List[int], tasks: List[int]) -> List[int]:
        n, m = len(servers), len(tasks)
        pq = []
        for i, s in enumerate(servers):
            heapq.heappush(pq, (s, i))
        
        ans = [-1] * m
        working = []
        cur_time = -1 # 现在的时间
        i = 0
        while i < m:
            cur_time += 1
            while working and working[0][0] <= cur_time:
                _, w, idx = heapq.heappop(working)
                heapq.heappush(pq, (w, idx))
            
            if pq: # 分配给哪台服务器
                while pq and cur_time >= i and i < m:
                    t = tasks[i]
                    w, idx = heapq.heappop(pq) # v权重，idx 第几台服务器
                    heapq.heappush(working, (cur_time + t, w, idx))
                    ans[i] = idx
                    i += 1
            else: # 没有服务器可以用,等待
                cur_time = working[0][0]-1 # 最先完成的服务器所用时间
        return ans
```
