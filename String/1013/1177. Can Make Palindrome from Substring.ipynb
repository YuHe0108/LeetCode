```python
from collections import Counter
from functools import lru_cache

class Solution:
    def canMakePaliQueries(self, s: str, queries):
        N = len(s)
        counters = [None] * (N + 1)
        counters[0] = Counter()
        for i, c in enumerate(s):
            idx = i + 1
            counter = counters[idx-1].copy()
            counter[c] += 1
            counters[idx] = counter
        
        @lru_cache(None)
        def helper(l, r, k):
            range_counter = counters[r + 1] - counters[l]
            odd_cnt = 0
            for cnt in range_counter.values():
                if cnt % 2 == 1:
                    odd_cnt += 1
            if (odd_cnt // 2) <= k:
                return True
        
        ans = [False] * (len(queries))
        for i, q in enumerate(queries):
            l, r, k = q
            if helper(l, r, k):
                ans[i] = True
        return ans
```
