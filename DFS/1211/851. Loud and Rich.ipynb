```python
from collections import defaultdict

class Solution:
    def loudAndRich(self, richer, quiet):
        n = len(quiet)
        ans = [x for x in range(n)]
        
        rich_dict = defaultdict(list)
        for x, y in richer:
            rich_dict[y].append(x)
        
        for i in range(n):
            if i not in rich_dict:
                continue
            stack = rich_dict[i][:]
            min_quiet = quiet[i]
            min_quiet_per = i
            seen = {i}
            while stack:
                k = stack.pop()
                if quiet[k] < min_quiet:
                    min_quiet = quiet[k]
                    min_quiet_per = k
                if k in rich_dict:
                    for j in rich_dict[k]:
                        if j not in seen:
                            stack.append(j)
                            seen.add(j)
            ans[i] = min_quiet_per
        return ans
```
